```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Lees de dataset in
eik = pd.read_csv('eik.csv', sep=';', decimal=',')

# Voeg een nieuwe kolom toe voor log(Volume)
eik['log_Volume'] = np.log(eik['Volume'])

# Selecteer de Hoogte en log(Volume) als de afhankelijke en onafhankelijke variabelen
X = eik['log_Volume'].values.reshape(-1, 1)
y = eik['Hoogte'].values.reshape(-1, 1)


# Maak een instantie van het lineaire regressiemodel
model = LinearRegression()

# Fit het model met de gegevens
model.fit(X, y)

# Haal de voorspelde waarden op
y_pred = model.predict(X)

# Maak een scatter plot van de gegevenspunten
plt.scatter(X, y, color='blue', label='Waargenomen data')

# Plot de regressielijn
plt.plot(X, y_pred, color='red', linewidth=2, label='Lineaire regressie')

# Voeg labels en titel toe
plt.xlabel('log(Volume)')
plt.ylabel('Hoogte')
plt.title('Lineaire Regressie: Hoogte voorspellen op basis van log(Volume)')

# Voeg een legenda toe
plt.legend()

# Toon de plot
plt.show()


# Haal de parameters van het regressiemodel op
intercept = model.intercept_[0]
slope = model.coef_[0][0]

# Bereken het R²-coëfficiënt en de gecorrigeerde R²-coëfficiënt
R2 = model.score(X, y)
n = len(eik)
adjusted_R2 = 1 - (1 - R2) * (n - 1) / (n - 2)

# Print de resultaten
print('Intercept:', intercept)
print('Slope (coëfficiënt):', slope)
print('R²-coëfficiënt:', R2)
print('Gecorrigeerde R²-coëfficiënt:', adjusted_R2)

```
